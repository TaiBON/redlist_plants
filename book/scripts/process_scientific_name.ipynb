```python
import psycopg2 as pg

import genlist_api                                                                                                                                                
g = genlist_api.Genlist() 
```

```python
DSN = 'dbname=nvdimp user=psilotum'
conn = pg.connect(DSN)
```

```python
QUERY = 'SELECT fullname FROM twredlist2017;'
with conn:
    with conn.cursor() as curs:                                                                                                                           
        curs.execute(QUERY)
        fullnames = curs.fetchall()
```

```python
for i in range(len(fullnames)):
    # replace double spaces
    if '  ' in fullnames[i][0]:
        replacedFullname = fullnames[i][0].replace('  ', ' ')
        UPDATE = '''
        UPDATE twredlist2017 SET
            fullname = '%s' 
        WHERE fullname = '%s'
        ''' % (replacedFullname, fullnames[i][0])
        print(replacedFullname)
        with conn:
            with conn.cursor() as curs:                                                                                                                           
                curs.execute(UPDATE)
    if 'ssp.' in fullnames[i][0]:
        replacedFullname = fullnames[i][0].replace('ssp.', 'subsp.')
        UPDATE = '''
        UPDATE twredlist2017 SET
            fullname = '%s' 
        WHERE fullname = '%s'
        ''' % (replacedFullname, fullnames[i][0])
        print(replacedFullname)
        with conn:
            with conn.cursor() as curs:                                                                                                                           
                curs.execute(UPDATE)
    # escape quote
    if "'" in fullnames[i][0]:
        replacedFullname = fullnames[i][0].replace("'", "''")
        UPDATE = '''
        UPDATE twredlist2017 SET
            fullname = '%s' 
        WHERE fullname = '%s'
        ''' % (replacedFullname, replacedFullname)
        print(replacedFullname)
        with conn:
            with conn.cursor() as curs:                                                                                                                           
                curs.execute(UPDATE)
    conn.commit()       

    if "\xa0" in fullnames[i][0]:
        print(fullnames[i][0].replace('\xa0', ' '))
        replacedFullname = fullnames[i][0].replace('\xa0', ' ')
        UPDATE = '''
        UPDATE twredlist2017 SET
            fullname = '%s' 
        WHERE fullname = '%s'
        ''' % (replacedFullname, fullnames[i][0])
        with conn:
            with conn.cursor() as curs:                                                                                                                           
                curs.execute(UPDATE)
            conn.commit()
```

```python
QUERY = 'SELECT fullname FROM twredlist2017;'
with conn:
    with conn.cursor() as curs:                                                                                                                           
        curs.execute(QUERY)
        fullnames = curs.fetchall()
for i in range(len(fullnames)):
    fname = fullnames[i][0]
    nameFormatted = g.fmtnameNew(fname, split=True, format_type='custom', 
                                 italic_b="", italic_e="")
    if "'" in fname:
        fname = fname.replace("'", "''")
    if "'" in nameFormatted[0]:
        nameFormatted[0] = nameFormatted[0].replace("'", "''")
    if "'" in nameFormatted[1]:
        nameFormatted[1] = nameFormatted[1].replace("'", "''")
    UPDATE = '''
        UPDATE twredlist2017 SET
            name = '%s', authors = '%s'
        WHERE fullname = '%s'
        ''' % (nameFormatted[0], nameFormatted[1], fname)
    with conn:
        with conn.cursor() as curs:                                                                                                                           
            curs.execute(UPDATE)
    print(nameFormatted)
conn.commit()
```
