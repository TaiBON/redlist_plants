```python
import csv
redlistRaw = []
with open('../../workspace/redlist171204.csv', 'r') as f:
    csvreader = csv.reader(f, delimiter='|', quotechar='"')
    for row in csvreader:
        print(row)
        redlistRaw.append(row)
```

```python
# SQL

import sqlite3
conn = sqlite3.connect('/tmp/test2.db')
```

```python
CREATE_DB = '''
CREATE TABLE redlist (
    id integer,
    family varchar,
    family_zh varchar,
    fullname varchar,
    zh_name varchar,
    category varchar
);
'''
curs = conn.cursor()
curs.execute(CREATE_DB)
conn.commit()
```

```python
for r in range(len(redlistRaw)):
    INSERT = '''
    INSERT INTO redlist (id,family,family_zh,fullname,zh_name,category)
    VALUES (%s, '%s', '%s', '%s', '%s', '%s');
    ''' % (redlistRaw[r][0], redlistRaw[r][3], redlistRaw[r][4],
           redlistRaw[r][5], redlistRaw[r][6], redlistRaw[r][10])
    print(INSERT)
    curs = conn.cursor()
    curs.execute(INSERT)
    conn.commit()
```

```python
QUERY = '''SELECT * FROM redlist WHERE family ='Acanthaceae'; '''
curs.execute(QUERY)
acanthaceae = curs.fetchall()
```
